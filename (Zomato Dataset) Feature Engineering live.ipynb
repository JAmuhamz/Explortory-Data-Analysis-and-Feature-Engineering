```python
df_country=pd.read_excel(r'C:\Users\USER\Desktop\projects\EDA and Feature Engineering\Country-Code.xlsx')
df_country
```

```python
#finding relationships between datasets
final_df=pd.merge(df, df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
#check data types

final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#PIE CHART- Top 3 countries

plt.pie(country_values[:3], labels=country_names[:3], autopct= '%1.1f%%')
plt.show()
```

OBSERVATIONS: Zomato max records or transactions are from India(94.4%) followed by United states(4.7%) and United kingdom(0.9%)

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
#change to a dataframe
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

OBSERVATIONS:
- When rating is between 4.5-4.9 = excellent
- when rating is between 4.0-4.4 = very good
- when rating is between 3.5-3.9 = good
- when rating is between 2.5-3.4 = average
- when rating is between 1.8-2.4 = poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(20,10)
sns.barplot(x='Aggregate rating',y="Rating count", data=ratings)
plt.show()
```

```python
matplotlib.rcParams['figure.figsize']=(20,10)
sns.barplot(x='Aggregate rating',y="Rating count",hue='Rating color',palette=['blue','red','orange','yellow','green','darkgreen'], data=ratings)
plt.show()
```

Observations:
- Not rated count is very high
- Max no. of rating is between 2.5 to 3.4

```python
##USE COUNTPLOT FOR CATEGORICAL VALUES
matplotlib.rcParams['figure.figsize']=(15,6)
sns.countplot(x='Rating color', data = ratings, palette =['blue','red','orange','yellow','green','darkgreen'])
plt.show()
```
